```python
import numpy
import math
%matplotlib inline
#%matplotlib notebook
import matplotlib.pyplot as plt
import scipy.stats
```

```python
xvals=numpy.linspace(0,1.0,num=1000)
#yvals=scipy.stats.uniform.pdf(xvals)
ytemp=numpy.random.uniform(0,1,1000)
yvals = -2*numpy.log(1-ytemp)
plt.figure()
plt.plot(xvals,yvals)
plt.ylim(0,max(yvals)*1.1)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("pdf of -2*ln(1 - uniform(0,1))")
plt.show()
```

Below is the qq plot of  -2*ln(1 - uniform(0,1)) against standard exponential distribution. 

```python
plt.figure()
scipy.stats.probplot(yvals, sparams=(), dist='expon', fit=True, plot=plt, rvalue=False)
plt.title("qq plot")
plt.show()
```
